```python
import subprocess

def run_external_script(file_path):
    """
    Runs an external Python script located at the given file path.
    :param file_path: The path to the Python script to run.
    """
    try:
        # 使用 subprocess 模块运行外部脚本
        completed_process = subprocess.run(["python", file_path], check=True, text=True, capture_output=True)
        
        # 打印输出结果
        print("STDOUT:", completed_process.stdout)
        print("STDERR:", completed_process.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error occurred while running the script: {e}")

if __name__ == "__main__":
    # 指定要运行的脚本的路径
    script_path = "AML_COURSEWORK_A_LOGISTIC.ipynb"  # 更改为实际脚本的路径
    
    # 调用函数运行脚本
    run_external_script(script_path)
```

```python
import subprocess
import nbformat
from nbconvert import PythonExporter

def convert_notebook_to_py(notebook_path, output_path):
    """
    Converts a Jupyter notebook to a Python script.
    :param notebook_path: The path to the Jupyter notebook.
    :param output_path: The path to save the converted Python script.
    """
    # 读取笔记本内容
    with open(notebook_path, 'r', encoding='utf-8') as nb_file:
        nb_content = nb_file.read()

    # 将笔记本内容转换为Python脚本
    nb_node = nbformat.reads(nb_content, as_version=4)
    exporter = PythonExporter()
    source, _ = exporter.from_notebook_node(nb_node)

    # 写入转换后的Python脚本
    with open(output_path, 'w', encoding='utf-8') as py_file:
        py_file.write(source)

def run_python_script(file_path):
    """
    Runs a Python script.
    :param file_path: The path to the Python script.
    """
    try:
        completed_process = subprocess.run(["python", file_path], check=True, text=True, capture_output=True)
        print("STDOUT:", completed_process.stdout)
        print("STDERR:", completed_process.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error occurred while running the script: {e}")

if __name__ == "__main__":
    # 指定 Jupyter 笔记本和输出 Python 脚本的路径
    notebook_path = r"C:\Users\001\AML_COURSEWORK_A_LOGISTIC.ipynb"  # 更改为你的 Jupyter 笔记本路径
    output_script_path = r"C:\Users\001\AML_COURSEWORK_A_LOGISTIC.py"  # 设置输出 Python 脚本的路径

    # 将笔记本转换为Python脚本
    convert_notebook_to_py(notebook_path, output_script_path)

    # 运行转换后的Python脚本
    run_python_script(output_script_path)
```

```python
import subprocess
import nbformat
from nbconvert import PythonExporter

def run_jupyter_notebook(notebook_path):
    """
    Runs a Jupyter notebook using nbconvert.
    :param notebook_path: The path to the Jupyter notebook.
    """
    try:
        # 使用 nbconvert 执行 Jupyter 笔记本
        subprocess.run(["jupyter", "nbconvert", "--to", "notebook", "--execute", notebook_path], check=True)
        print(f"Successfully ran {notebook_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error occurred while running the notebook: {e}")

def main():
    # 程序列表
    programs = {
        "1": "C:\\Users\\001\\AML_COURSEWORK_A_LOGISTIC.ipynb",
        "2": "/path/to/your/script2.py",
        # 在这里添加更多程序
    }

    # 打印菜单
    print("Please select a program to run:")
    for key, value in programs.items():
        print(f"{key}: {value}")

    # 用户选择
    choice = input("Enter your choice: ")

    # 根据用户选择运行对应的程序
    if choice in programs:
        print(f"Running {programs[choice]}...")
        if programs[choice].endswith('.ipynb'):
            run_jupyter_notebook(programs[choice])
        else:
            run_python_script(programs[choice])
    else:
        print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
import subprocess
import nbformat
from nbconvert import PythonExporter

def convert_notebook_to_py(notebook_path, output_path):
    """
    Converts a Jupyter notebook to a Python script.
    :param notebook_path: The path to the Jupyter notebook.
    :param output_path: The path to save the converted Python script.
    """
    with open(notebook_path, 'r', encoding='utf-8') as nb_file:
        nb_content = nb_file.read()

    nb_node = nbformat.reads(nb_content, as_version=4)
    exporter = PythonExporter()
    source, _ = exporter.from_notebook_node(nb_node)

    with open(output_path, 'w', encoding='utf-8') as py_file:
        py_file.write(source)

def run_python_script(file_path):
    """
    Runs a Python script.
    :param file_path: The path to the Python script.
    """
    try:
        completed_process = subprocess.run(["python", file_path], check=True, text=True, capture_output=True)
        print("STDOUT:", completed_process.stdout)
        print("STDERR:", completed_process.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error occurred while running the script: {e}")

def display_menu(options):
    """
    Displays a menu of options and gets the user's choice.
    :param options: A dictionary of option number to file path.
    :return: The chosen file path or None if invalid choice.
    """
    print("Please select a program to run:")
    for key, value in options.items():
        print(f"{key}: {value}")

    choice = input("Enter your choice: ")
    return options.get(choice, None)

if __name__ == "__main__":
    # 菜单选项，键为菜单选项，值为对应脚本的路径
    menu_options = {
        "1": r"C:\\Users\\001\\AML_COURSEWORK_A_LOGISTIC.ipynb",
        "2": "/path/to/your/notebook2.ipynb",
        # 可以在此处添加更多选项
    }

    # 显示菜单并获取用户选择
    chosen_file = display_menu(menu_options)

    if chosen_file:
        if chosen_file.endswith('.ipynb'):
            # 转换并运行 Jupyter 笔记本
            py_script_path = r"C:\\Users\\001\\AML_COURSEWORK_A_LOGISTIC.py"  # 设置输出 Python 脚本的路径
            convert_notebook_to_py(chosen_file, py_script_path)
            run_python_script(py_script_path)
        else:
            # 直接运行 Python 脚本
            run_python_script(chosen_file)
    else:
        print("Invalid choice. Please select a valid option.")
```
